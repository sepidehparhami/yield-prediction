```python
n_molecules = yields.shape[0]
n_coords = 26999
```

```python
n_molecules
```

```python
# initialize a data frame and fill in ESPs from file
# later we will save the (x,y,z) coordinates too to make use of them
data = pd.DataFrame(index = ligands, columns=range(n_coords))


directory = './data' 
      
#loop through files in directory
for filename in os.listdir(directory):
    data_point = pd.read_csv('./data/{}'.format(filename))
    data_point.columns = ('x', 'y', 'z', 'esp')
    name_parts = filename.split("_")

        
    #extract the ligand name from the filename
    if name_parts[1][0] == "T":
        ligand = name_parts[1][:4]
    else:
        ligand = name_parts[1][:3]
        
    #add label and data point to their respective arrays 
    data.loc[ligand] = data_point['esp']
```

```python
data
```

```python
#split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(data, yields, test_size=0.2, random_state=42, shuffle=True)
```

```python
#create and fit linear regression model
LR = LinearRegression()
LR.fit(x_train, y_train)
y_prediction =  LR.predict(x_test)
```

```python
print('MSE:', mean_squared_error(y_test, y_prediction))
print('R^2:', r2_score(y_test, y_prediction))
```
